```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dropout, Dense
from tensorflow.keras.optimizers import Adam

# --- 1) Load & preprocess the data --- #
path = r"C:\Users\91939\OneDrive\Documents\my work\DATA SCIENCE-AI&ML\smart-market\datasets\Google_stock_price.csv"
df = pd.read_csv(path, parse_dates=["Date"])          # read CSV
df.set_index("Date", inplace=True)                     # set date index
df.sort_index(inplace=True)                            # ensure chronological order
df.ffill(inplace=True)                                 # forward-fill any gaps

# --- 2) Scale the 'Close' column to [0,1] --- #
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df["Close_Scaled"] = scaler.fit_transform(df[["Close"]])  # now df has Close_Scaled

# --- 3) Sequence‐creation helper function --- #
def make_sequences(data, window_size=50):
    """
    Given a 1D array, returns:
      X: array of shape (n_samples, window_size)
      y: array of shape (n_samples,)
    """
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i : i + window_size])
        y.append(data[i + window_size])
    return np.array(X), np.array(y)

# Build full dataset of sequences
X_all, y_all = make_sequences(df["Close_Scaled"].values, window_size=50)

# Reshape X_all for LSTM: (samples, timesteps, features)
X_all = X_all.reshape(X_all.shape[0], X_all.shape[1], 1)

# --- 4) Train/validation split --- #
X_train, X_val, y_train, y_val = train_test_split(
    X_all, y_all, test_size=0.2, random_state=42
)

# --- 5) Model factory function --- #
def create_model(dropout_rate=0.2, learning_rate=0.001):
    model = Sequential()
    model.add(Input(shape=(X_train.shape[1], X_train.shape[2])))
    model.add(LSTM(50, return_sequences=True))
    model.add(Dropout(dropout_rate))
    model.add(LSTM(50))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))
    model.compile(optimizer=Adam(learning_rate), loss="mean_squared_error")
    return model

# --- 6) Hyperparameter grid definitions --- #
batch_sizes    = [32, 64]
epochs_list    = [50, 100]
dropout_rates  = [0.2, 0.5]
learning_rates = [0.001, 0.01]

# --- 7) Manual grid‐search loop --- #
best_score, best_params, best_model = float("inf"), None, None

for bs in batch_sizes:
    for ep in epochs_list:
        for dr in dropout_rates:
            for lr in learning_rates:
                print(f"Training bs={bs}, epochs={ep}, dropout={dr}, lr={lr}")
                model = create_model(dropout_rate=dr, learning_rate=lr)
                history = model.fit(
                    X_train, y_train,
                    validation_data=(X_val, y_val),
                    epochs=ep,
                    batch_size=bs,
                    verbose=0
                )
                val_loss = min(history.history["val_loss"])
                print(f" → val_loss={val_loss:.4f}")
                if val_loss < best_score:
                    best_score  = val_loss
                    best_params = {"batch_size": bs, "epochs": ep, "dropout_rate": dr, "learning_rate": lr}
                    best_model  = model

print("\nBest params:", best_params)
print("Best validation loss:", best_score)
```
